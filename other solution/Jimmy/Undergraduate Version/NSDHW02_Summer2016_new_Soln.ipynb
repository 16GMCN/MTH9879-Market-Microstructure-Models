### 1. (10 points)

Let $\sigma(t)$ and $r(t)$ be positive deterministic functions of $t$. Define

$$
  X_t = \int_0^t r(s) ds + \int_0^t \sigma(s) dB_s, \quad S_t = s_0 e^{X_t - \frac12 \int_0^t \sigma^2(s) ds}.
$$

(a) What are the distributions of $X_t$ and $S_t$?

(b) Determine the mean and variance of $X_t$ and $S_t$.

### <font color=blue> Solution </font>

(a) Since $r(s)$ and $\sigma(s)$ are both deterministic, 

$$X_t \sim N\left(\int_0^t r(s)ds, \int_0^t \sigma^2(s) ds \right). $$ 

In other words, $X_t$ is normally distributed with mean $\int_0^t r(s) ds$ and variance $\int_0^t \sigma^2(s) ds$. 
For $S_t$, since $\log S_t = X_t -\frac12 \int_0^t \sigma^2(s)ds$, 

$$
S_t \sim \log N\left(\int_0^t r(s)ds, \int_0^t \sigma^2(s) ds \right).
$$

In other words, $S_t$ is lognormally distributed with the same parameters as for $X_t$. 

(b) From (a), we have immediately

\begin{eqnarray*}
&& \mathbb{E}[X_t] = \int_0^t r(s) ds, \\
&& \mbox{Var}[X_t] = \int_0^t \sigma^2(s)ds. 
\end{eqnarray*}

For $S_t$, recall that if $X \sim \log N(\mu, \sigma^2)$, then $\Eof{X} = e^{\mu + \frac{\sigma^2}2}$ and $\mbox{Var}[X] = \left( e^{\sigma^2}- 1\right) e^{2\mu + \sigma^2}$. 
Thus,
\begin{eqnarray*}
&& \Eof{S_t} = e^{R + \frac\Sigma2}, \\
&& \mbox{Var}[S_t] = \left( e^{\Sigma} - 1\right) e^{2R + \Sigma}
\end{eqnarray*}

where $R = \int_0^t r(u)du$ and $\Sigma = \int_0^t \sigma^2(u)du$. 

### 2. (10 points)

Let $X$ be the random variable given by the stochastic integral 

$$
X = \int_a^b f_s dB_s
$$

where $f_s$ is an adapted stochastic process. 
Recall that Ito isometry states 

$$
\Eof{X^2} = \int_a^b \Eof{f_s^2} ds.
$$

Use Ito isometry to determine the variance of the following random variables. 

(a) $X = \int_0^t B_s^2 dB_s$

(b) $X = \int_a^b \sqrt t e^{B_t} dB_t$

(c) $X = \int_a^b \sqrt t \sin(B_t) dB_t$

### <font color=blue> Solution.</font>

Note that in all case we have $\Eof{X} = 0$ since $X$ is given by a stochastic integral. Thus, $\mbox{Var}[X] = \Eof{X^2}$.

<font color=blue> (a). </font> By applying Ito isometry we obtain

\begin{eqnarray*}
&& \Eof{X^2} = \int_0^t \Eof{B_s^4} ds \\
&=& \int_0^t 3 s^2 ds \quad (\because B_s \sim N(0, s)) \\
&=& t^3.
\end{eqnarray*}

<font color=blue> (b). </font> Similarly, by applying Ito isometry we obtain

\begin{eqnarray*}
&& \Eof{X^2} = \int_a^b \Eof{t e^{2 B_t}} dt \\
&=& \int_a^b t e^{2t}  dt \quad (\because 2 B_t \sim N(0, 4t)) \\
&=& \left. \left( \frac t2 e^{2t} - \frac14 e^{2t}\right)\right|_a^b 
= \frac12 \left( b - \frac12 \right)e^{2b} - \frac12 \left(a - \frac12 \right)e^{2a}.
\end{eqnarray*}

<font color=blue> (c). </font> Again, by applying Ito isometry we obtain

\begin{eqnarray*}
&& \Eof{X^2} = \int_a^b \Eof{t \sin^2(B_t)} dt. 
\end{eqnarray*}

We calculate $\Eof{\sin^2(B_t)}$ as follows. Recall that if $X \sim N(0,1)$, the characteristic function $\phi_X$ of $X$ is given by 

$$
\phi_X(\xi) := \Eof{e^{i\xi X}} = \Eof{\cos(\xi X)} + i \Eof{\sin(\xi X)} = e^{-\frac{\xi^2}2}.
$$

Hence, $\Eof{\cos(\xi X)} = e^{-\frac{\xi^2}2}$. It follows that 

\begin{eqnarray*}
&& \Eof{\sin^2(B_t)} = \frac12 \Eof{1 - \cos(2 B_t)} \\
&=& \frac12 - \frac12 \Eof{\cos(2\sqrt t B_1)} \quad (\because B_t \sim \sqrt t B_1 \sim N(0,1)) \\
&=& \frac12 - \frac12 e^{-2t}.
\end{eqnarray*}

Thus, 

\begin{eqnarray*}
&& \Eof{X^2} = \int_a^b \Eof{t \sin^2(B_t)} dt \\
&=& \int_a^b t \left(\frac12 - \frac12 e^{-2t}\right) dt \\
&=& \frac14(b^2 - a^2) - \frac14\left( b + \frac12 \right) e^{-2b} + \frac14\left( a + \frac12 \right) e^{-2a}.
\end{eqnarray*}

### 3. (6 points)

Define the process $X_t$ as $X_t = B_t - t B_1$, for $t \in [0,1]$. 

For any $0 \leq t_1 \leq t_2 \leq 1$, 

(a) determine the quantities $\Eof{X_{t_1}}$, $\Eof{X_{t_2}}$, $\mbox{Var}[X_{t_1}]$, $\mbox{Var}[X_{t_2}]$, and $\mbox{cov}[X_{t_1}, X_{t_2}]$. 

(b) what is the joint probability density function $\Pof{X_{t_1} = x_1, X_{t_2} = x_2}$?



### <font color=blue> Solution.</font>

<font color=blue>(a).</font> $\Eof{X_{t_1}} = \Eof{B_{t_1} - t_1 B_1} = \Eof{B_{t_1}} - t_1 \Eof{B_1} = 0$. Similarly, we have $\Eof{X_{t_2}} = 0$. 
For $\mbox{Var}[X_{t_1}] = \mbox{Var}[B_{t_1} - t_1 B_1] = \mbox{Var}[B_{t_1}] + t_1^2 \mbox{Var}[B_1] - 2 t_1 \mbox{cov}[B_{t_1}, B_1] = t_1 + t_1^2 - 2 t_1^2 = t_1 (1 - t_1)$. Similarly, one can show that $\mbox{Var}[X_{t_2}] = t_2(1 - t_2)$. 

Finally, we calculate $\mbox{cov}[X_{t_1}, X_{t_2}]$ as follows. Assume $t_1 < t_2$, 

\begin{eqnarray*}
&& \mbox{cov}[X_{t_1}, X_{t_2}] = \mbox{cov}[B_{t_1} - t_1 B_1, B_{t_2} - t_2 B_1] \\
&=& \mbox{cov}[B_{t_1}, B_{t_2}] - t_2 \mbox{cov}[B_{t_1}, B_1] - t_1 \mbox{cov}[B_1, B_{t_2}] + t_1 t_2 \mbox{Var}[B_1] \\
&=& t_1 - t_2 t_1 - t_1 t_2 + t_1 t_2 = t_1(1 - t_2).
\end{eqnarray*}

<font color=blue>(b).</font> Note that, since $X_{t_1}$ and $X_{t_2}$ are jointly normal with mean $(0,0)$ and covariance matrix given in (a), the joint pdf is given by 

$$
p(x_1, x_2) = \frac1{2\pi |\Sigma|^{\frac12}} e^{-\frac12 \boldsymbol x' \Sigma^{-1} \boldsymbol x},
$$

where $\boldsymbol x = (x_1, x_2)$, $\Sigma$ is the covariance matrix

$$
\Sigma = \left[\begin{array}{cc}
t_1(1 - t_1) & t_1 (1 - t_2) \\
t_1 (1 - t_2) & t_2 (1 - t_2)
\end{array}\right]
$$

and $|\Sigma|$ is the determinant of $\Sigma$.
