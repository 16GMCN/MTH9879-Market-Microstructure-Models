### 1. (10 points)

Let $W_t$ be a Brownian motions. 

(a) Solve the SDE for geometric Brownian motion or Black-Scholes model

$$
dX_t = \mu X_t dt + \sigma X_t dW_t \quad \Longleftrightarrow \quad \frac{dX_t}{X_t} = \mu dt + \sigma dB_t, \quad X_0 = x
$$

by applying Ito's formula to $\log X_t$.

(b) Solve the SDE for Ornstein-Uhlenbeck process or the Vasicek model

$$
dX_t = \lambda(m - X_t) dt + \sigma dW_t, \quad X_0 = x
$$

by applying Ito's formula to $e^{\lambda t} X_t$. Determine the mean and variance of $X_t$.  


#### <font color=blue> 1(a) Solution </font>

Note that the quadratic variation of $X_t$ is 

$$
[X]_t = \left[\int_0^t \sigma X_s dB_s \right] = \int_0^t \sigma^2 X_s^2 ds \quad \rightsquigarrow \quad
d[X]_t = \sigma^2 X_t^2 dt
$$

Let $f(x) = \log x$, apply Ito's formula to $\log X_t$ we have

\begin{eqnarray*}
&& d\log X_t = df(X_t) = f'(X_t) dX_t + \frac12 f''(X_t) d[X]_t \\
&=& \frac{dX_t}{X_t} - \frac12 \frac{d[X]_t}{X_t^2} = \mu dt + \sigma dB_t - \frac{\sigma^2}2 dt \\
&=& \left( \mu - \frac{\sigma^2}2 \right) dt + \sigma dB_t.
\end{eqnarray*}

Hence, by integrating from $0$ to $T$ on both sides of the last equation we obtain 

\begin{eqnarray*}
&& \log X_T - \log X_0 = \left(\mu - \frac{\sigma^2}2\right) T + \sigma B_T \quad
\Longrightarrow \quad X_T = X_0 e^{\left(\mu - \frac{\sigma^2}2\right) T + \sigma B_T}. 
\end{eqnarray*}

#### <font color=blue> 1(b) Solution </font>

We solve the SDE by applying Ito's formula to $e^{\lambda t} X_t$ as follows. Let $f(t,x) = e^{\lambda t} x$. Note that $d[X]_t = \sigma^2 dt$.

\begin{eqnarray*}
 d(e^{\lambda t} X_t) &=& d f(t,X_t) \\
&=& f_t(t,X_t) dt + f_x(t,X_t) dX_t + \frac12 f_{xx}(t,X_t) d[X]_t \\
&=& \lambda e^{\lambda t} X_t dt + e^{\lambda t} dX_t \\
&=& \lambda e^{\lambda t} X_t dt + e^{\lambda t} \left[ \lambda(m - X_t) dt + \sigma dB_t \right] \\
&=& \lambda e^{\lambda t} m  dt + \sigma e^{\lambda t} dB_t
\end{eqnarray*}

Hence, by integrating from $0$ to $T$ on both sides of the last equation we obtain  

\begin{eqnarray*}
 && e^{\lambda T} X_T - X_0 = \int_0^T e^{\lambda t} m \lambda  dt + \int_0^T \sigma e^{\lambda t} dB_t \\
 &\Longrightarrow& X_T = e^{-\lambda T} x + \left(1 - e^{-\lambda T} \right) m + \sigma \int_0^T e^{-\lambda (T-t)} dB_t.
\end{eqnarray*}

Hence, $X_T$ is normally distributed with mean $e^{-\lambda T} x + \left(1 - e^{-\lambda T} \right) m$ and variance $\sigma^2 \int_0^T e^{-2\lambda (T-t)} dt$.

### 2. (10 points)

Let $u$ be a solution to the terminal value problem

\begin{eqnarray*}
  && u_t + \frac{\sigma^2 x^2}2 u_{xx} + r x u_x = r u + D, \quad t < T,  \\
  && u(x,T) = x,
\end{eqnarray*}

where $\sigma$, $r$, and $D$ are constants. 

(a) Find a stochastic representation for $u$, clearly indicate what the underlying SDE is and which expectation to evaluate. 

(b) Use the result from 1(a) to determine the value $u(x,0)$.

#### <font color=blue> 2(a) Solution </font>

$u(x,t)$ has the stochastic representation

\begin{eqnarray*}
u(x,t) &=& \E\left[\left. e^{-\int_t^T r ds} X_T - \int_t^T e^{-\int_t^\tau rds} D d\tau \right|X_t = x\right] \\
&=& e^{-r(T-t)} \E[X_T|X_t = x]- \frac Dr \left[1 - e^{-r(T-t)} \right],
\end{eqnarray*}

where $X_t$ satisfies the SDE $dX_t = \sigma X_t dB_t + r X_t dt$. $B_t$ is a Brownian motion. 

#### <font color=blue> 2(b) Solution </font>

From 2(a) we have 

$$
X_t = X_0 e^{rT - \frac{\sigma^2T}2 + \sigma B_T}.
$$

Note that, since $B_T$ is normally distributed with mean 0 and variance $T$, $\E\left[e^{\sigma B_T}\right] = e^{\frac{\sigma^2T}2}$. Therefore,  

$$
\E[X_T|X_0 = x] = \E\left[\left.X_0 e^{rT - \frac{\sigma^2T}2 + \sigma B_T} \right| X_0 = x\right]
= x e^{rT - \frac{\sigma^2T}2} \E\left[e^{\sigma B_T}\right] = x e^{rT}.
$$

Finally, 

\begin{eqnarray*}
u(x,0) &=& e^{-rT} \E[X_T|X_0 = x]- \frac Dr \left[1 - e^{-rT} \right] = x - \frac Dr \left[1 - e^{-rT} \right].  
\end{eqnarray*}


## The Parlour (1998) model

### 3. (10 points) 
Consider the first step $t = 1$ of the 3-step order book case presented in the lecture slides. Suppose that the order book at (or just before) $t = 1$ has one share offered and one share bid ($n_B = n_A = 1$).

Give the probabilities as of a market buy, market sell, limit buy and limit sell at $t = 1$.

### <font color=blue> Solution. </font>

The order book at time $t=1$ is $(1,1)$ so submitting LS will cause the book at time $t=2$ to be $(1,2)$.  In order for a new limit sell order to be filled, there needs to be both a MB order at $t=2$ and also a market buy order at $t=3$.

The next trader, if a buyer, will see one share already in the queue and will not submit LB; he will choose between doing nothing and MB.  The probability of MB (as already computed) is $5/22$ per period.  The probability of two MB's is thus

$$
P_{fill}=5/22 \times 5/22.
$$

As per Table 3, a seller chooses between

Strategy | Payoff | Numerical 
---| --- | ---
Market sell (MS) | $B-\beta_t\,V$ | $5-\beta\,5.5$ 
Limit sell (LS) | $(A-\beta_t\,V)\,P_{fill}$ | $(6-\beta\,5.5)\times(5/22)^2$
Do nothing | 0 | 0

MS and LS give equal benefit if

$$
(A-\beta\,V) \times P_{fill}=B-\beta\,V
$$

The break-even $\beta$ is then given by

$$
\beta^\star = \frac{B-A\,P_{fill}}{V\,(1-P_{fill})} =   \frac{5-6\,(5/22)^2}{5.5\,(1-(5/22)^2)} = 0.899.
$$

Doing nothing is optimal if $(A-\beta\,V)\,P_{fill} < 0$ so do nothing if
$$
\beta > \frac{A}{V} = \frac{12}{11} = 1.091.
$$

Thus, conditional on being a seller, actions are as follows:

* MS if $\beta < 0.899$

* LS if $0.899 \leq \beta < 1.091$.

* Do nothing if $1.091 \leq \beta < 2 $.

The relative probabilities conditional on being a seller are just half of the $\beta$.  And the probability of being a seller is $1/2$.  So

$$
\PP_{MS}=0.899/2/2=0.225;\,\PP_{LS}=(1.091-0.899)/2=0.048;
$$

By symmetry

$$
\PP_{MB}=\PP_{MS}=0.225;\,\PP_{LB}=\PP_{LS}=0.048;
$$

so

$$
\PP_{\text{Do nothing}}=1-2\,(\PP_{MS}+\PP_{LS})=0.455
$$

### 4. (4 points) 
Explain why the optimal market/limit order decision depends in general on quantity available on both sides of the order book in the Parlour model.

### <font color=blue>Solution</font>.
Wlog consider the situation of a buyer. The greater the quantity on the bid side of the
book, the farther behind he will be in the queue: He will be less likely to place LB and more
likely to place MB. The greater the quantity on the offered (ask) side of the book, the more
likely it is that a subsequent seller will place a MS order so the more likely it is that a LB
order will be filled: He will be more likely to place LB and less likely to place MB.
