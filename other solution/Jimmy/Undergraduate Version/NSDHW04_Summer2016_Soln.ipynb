## The Kyle (1985) model

### 1. (14 points)

Suppose that the monopolistic trader in the Kyle (1985) model is not told the true value $\tilde S$ but is instead given a signal $x=\tilde S+\epsilon$ with $\epsilon \sim N(0,\sigma_\epsilon^2)$, where $\tilde S$ and $\epsilon$ are assumed independent.

(a) Compute the conditional expectation $\mathbb{E}[\tilde S|x]$.

(b) If $\mathcal I$ assumes that the market maker $\mathcal M$ generates a price function of the form $p=\mu +\lambda\,y$, show that the informed trader's optimal (profit maximizing) demand is given by

$$
y_i=\frac{\phi}{2\,\lambda}\,\left(x-\mu\right)
$$

where

$$
\phi = \frac{\sigma^2}{\sigma^2+\sigma_\epsilon^2}.
$$

(c) Following the derivation from the lecture slides, use the result of (b) to deduce equation (11) in the lecture slides:

$$
\lambda=\frac{\sigma}{2\,\sigma_u}\,\sqrt{\phi}.
$$


### <font color=blue> Solution. </font>

(a) Applying Bayes' Rule, the conditional expectation of $\tilde S$ is the precision-weighted sum:

$$
\EE[\tilde S|x]=\frac{\tau_S}{\tau_S+\tau_x}\,\mu+\frac{\tau_x}{\tau_S+\tau_x}\,x=(1-\phi)\,\mu+\phi\,x
$$

where $\tau_x=1/\sigma_x^2$ and $\tau_S=1/\sigma^2$.

(b) $\cI$ maximizes his expected profit $\pi$ conditional on his signal $x$ assuming $\cM$ will set $p=\mu+\lambda\,y$ for some $\lambda$.

$$
\pi=\EE\left[y_i\,(\tilde S-p)|x\right]=y_i\,\left(\EE[\tilde S|x]-\mu-\lambda\,y_i\right)
$$

so his optimal demand $y_i$ is given by

$$
y_i=\frac{\EE[\tilde S|x]-\mu}{2\,\lambda}.
$$

It follows that

$$
y_i=\frac{\phi}{2\,\lambda}(x-\mu)
$$

as required.

(c) Now the rest of the derivation proceeds exactly as before.  $\cM$ conjectures that $\cI$'s demand function is of the form $y_i=a+b\,x$.

Knowing the optimization process followed by $\cI$, $\cM$ can solve for $a$ and $b$:

$$
y_i=\frac{\phi}{2\,\lambda}(x-\mu)=a+b\,x
$$

So

$$
a= -\frac{\mu\,\phi}{2\,\lambda} \text { and } b= \frac{\phi}{2\,\lambda}.
$$

$\cM$ sets the pricing function as

$$
p(y)=\mu+\lambda\,y=\EE[\tilde S|y]
$$

Recall that $y=y_i+y_u=a+b\,x+y_u=a+b\,(\tilde S +\epsilon)+y_u$ so

$$
\Var[y]=b^2\,\sigma_x^2+\sigma_u^2=b^2\,(\sigma^2+\sigma_\epsilon^2)+\sigma_u^2;\,\Cov[\tilde S,y]=b\,\sigma^2;
$$

Then using equation (11) in the lecture to regress $\tilde S$ against $y$ gives

$$
\lambda = \frac{\Cov[\tilde S,y]}{\Var[y]}=\frac{b\,\sigma^2}{b^2\,(\sigma^2+\sigma_\epsilon^2)+\sigma_u^2}
$$

Also, recall that $\cM$'s conjecture about $\cI$'s demand gives

$$
b=\frac{\phi}{2\,\lambda}
$$

Solving for $b$ and $\lambda$ gives

$$
b=\sqrt{\phi}\,\frac{\sigma_u}{\sigma};\, \lambda=\frac{\sigma}{2\,\sigma_u}\,\sqrt{\phi}
$$

### 2. (6 points) 

Suppose that a broker $\mathcal P$ spots an insider and decides to submit his own piggyback orders.  Specifically, if $\mathcal I$ submits demand $y_i$, $\mathcal P$ submits demand $\gamma\,y_i$ with $\gamma>0$.

(a) Assuming that $\mathcal I$ can detect how much piggyback order flow there is, deduce that the equilibrium Kyle lambda is unchanged from the original case where there are no piggyback orders.

(b) Derive expressions for $\mathcal I$'s optimal demand $y_i$ and his expected P&L.

(c) Explain to a non-expert what happens to your trade capacity and your profitability if someone is piggybacking on your trades.

### <font color=blue> Solution. </font>

This is equivalent to the original problem except that $\mathcal I$ and $\mathcal P$ now act as one informed trader $\mathcal I$'s.  Thus $\lambda$ is unchanged.  The new optimal demand for $\mathcal I$ is

$$
y_i=\frac{1}{1+\gamma}\,\frac{\tilde S-\mu}{2\,\lambda}
$$

and his expected profit is likewise reduced by a factor $1+\gamma$.  In words, if somebody else is piggybacking on your trades, both your trade capacity and your profitability are reduced by a factor.
