### 1. (10 points)

Let $W_t$ be a Brownian motions. 

(a) Solve the SDE for geometric Brownian motion or Black-Scholes model

$$
dX_t = \mu X_t dt + \sigma X_t dW_t \quad \Longleftrightarrow \quad \frac{dX_t}{X_t} = \mu dt + \sigma dB_t, \quad X_0 = x
$$

by applying Ito's formula to $\log X_t$.

(b) Solve the SDE for Ornstein-Uhlenbeck process or the Vasicek model

$$
dX_t = \lambda(m - X_t) dt + \sigma dW_t, \quad X_0 = x
$$

by applying Ito's formula to $e^{\lambda t} X_t$. Determine the mean and variance of $X_t$.  


### The answer to question(a) is as follows.
Define $f(x)\equiv lnx$. Then we have
$$
f'(x)=\frac1x,\ f''(x)=-\frac1{x^2}
$$
Apply Ito formula to $f(X_t)$, we can get

\begin{eqnarray}
d(lnX_t) && =\frac1{X_t}dX_t-\frac12\frac1{X_t^2}d[X]_t\\
&& =\frac1{X_t}(\mu dt+\sigma dW_t)X_t-\frac1{2X_t^2}\sigma^2X_t^2dt\\
&& =(\mu-\frac12\sigma^2)dt+\sigma dW_t
\end{eqnarray}

Taking integral on both sides yields
$$
lnX_t=lnx+(\mu-\frac12\sigma^2)t+\sigma W_t
$$
or
$$
X_t=x\cdot exp[(\mu-\frac12\sigma^2)t+\sigma W_t]
$$
which is indeed a geometric BM.

### The answer to question(b) is as follows.
Define $f(x,t)\equiv e^{\lambda t}x$, then we have
$$
f_t=\lambda e^{\lambda t}x,\ \ f_x=e^{\lambda t},\ \ f_{xx}=0
$$
Apply Ito formula to $f(X_t)$, we can get

\begin{eqnarray}
d(e^{\lambda t}X_t)&&=\lambda e^{\lambda t}X_tdt+e^{\lambda t}dX_t\\
&&=\lambda e^{\lambda t}X_tdt+e^{\lambda t}\lambda(m-X_t)dt+e^{\lambda t}\sigma dW_t\\
&&=e^{\lambda t}\cdot\lambda m\cdot dt+e^{\lambda t}\sigma dW_t
\end{eqnarray}

Taking integral on both sides yields
\begin{eqnarray}
e^{\lambda t}X_t&&=\lambda m\int_0^te^{\lambda s}ds+\sigma\int_0^te^{\lambda s}dW_s+x\\
&&=m(e^{\lambda t}-1)+\sigma\int_0^te^{\lambda s}dW_s+x
\end{eqnarray}

or
$$
X_t=xe^{-\lambda t}+m(1-e^{-\lambda t})+\sigma e^{-\lambda t}\int_0^te^{\lambda s}dW_s
$$

Now let's discuss the mean and variance of $X_t$.  
It's clear that
$$
M_t\equiv\int_0^te^{\lambda s}dW_s
$$
is a martingale. So $EM_t=0$. Thus
$$
EX_t=xe^{-\lambda t}+m(1-e^{-\lambda t})
$$
\begin{eqnarray}
VarX_t&&=\sigma^2e^{-2\lambda t}\int_0^te^{2\lambda s}ds\ \ (Ito\ Isometry)\\
&&=\frac{\sigma^2}{2\lambda}(1-e^{-2\lambda t})
\end{eqnarray}

### 2. (10 points)

Let $u$ be a solution to the terminal value problem

\begin{eqnarray*}
  && u_t + \frac{\sigma^2 x^2}2 u_{xx} + r x u_x = r u + D, \quad t < T,  \\
  && u(x,T) = x,
\end{eqnarray*}

where $\sigma$, $r$, and $D$ are constants. 

(a) Find a stochastic representation for $u$, clearly indicate what the underlying SDE is and which expectation to evaluate. 

(b) Use the result from 1(a) to determine the value $u(x,0)$.

### The answer to question(a) is as follows.
According to the ultimate version of Feynman-Kac Theorem (i.e., the content of "Backward second order parabolic PDEs"), we can conclude that

\begin{eqnarray}
u(x,t)&&=E_{t,x}[e^{-\int_t^Trds}\cdot X_T-\int_t^Te^{-\int_t^{\tau} rds}Dd\tau]\\
&&=E_{t,x}[e^{-r(T-t)}\cdot X_T-D\int_t^Te^{-r(\tau-t)}d\tau]\\
&&=E_{t,x}[e^{-r(T-t)}\cdot X_T+\frac Dr(e^{-r(T-t)}-1)]
\end{eqnarray}
where
$$
dX_t=rX_tdt+\sigma X_tdW_t
$$
and
$$
E_{t,x}[\cdot]\equiv E[\cdot|X_t=x]
$$

### The answer to question(b) is as follows.
Using the result of question(a), we have
\begin{eqnarray}
u(x,0)&&=E_{t,x}[e^{-rT}\cdot X_T+\frac Dr(e^{-rT}-1)]\\
&&=e^{-rT}E[X_T|X_t=x]+\frac Dr(e^{-rT}-1)
\end{eqnarray}

Now the only work left is to get the value of $E[X_T|X_t=x]$.\\
With the help of the result of Q1(a), we have
\begin{eqnarray}
E(X_T|X_t=x)&&=x\cdot E[exp(r-\frac12\sigma^2)(T-t)+\sigma(W_T-W_t)]\\
&&=x\cdot exp[(r-\frac12\sigma^2)(T-t)]\cdot E[exp(\sigma(W_T-W_t))]\\
&&=x\cdot exp[(r-\frac12\sigma^2)(T-t)]\cdot exp(\frac{\sigma^2}2(T-t))\\
&&=x\cdot exp[r(T-t)]
\end{eqnarray}
In the last step, we use the normal distribution of BM increments and log-normal random variable expectation formula.  
So in conclusion,
\begin{eqnarray}
u(x,0)&&=e^{-rT}x\cdot exp[rT]+\frac Dr(e^{-rT}-1)\\
&&=x+\frac Dr(e^{-rT}-1)
\end{eqnarray}

## The Bouchaud-M&eacute;zard-Potters order book approximation

### 3. (16 points)

When $\lambda(u)=\alpha$,  a constant, the BMP order density becomes

\begin{eqnarray*}
\rho_{BMP}(\Delta) &=& \frac{\alpha}{\delta}\,\left\{e^{-\Delta/{\tilde \Delta}}\,\int_0^{\Delta/{\tilde\Delta}}\,\sinh(u)\,du + \sinh(\Delta/{\tilde \Delta})\,\int_{\Delta/{\tilde\Delta}}^\infty\,e^{-u}\,du\right\} \\
&=&\frac{\alpha}{\delta}\,\left\{1 - e^{-\Delta/{\tilde \Delta}}  \right\}
\end{eqnarray*}

where $\tilde \Delta$ denotes the characteristic distance defined by

$$
\tilde \Delta = \sqrt{\frac{\sigma^2}{2\,\delta}}.
$$

### The answer to question(d) is as follows.
We know that
$$
\alpha=1,\ \mu=10,\ \delta=0.2
$$
So, using the formula in the question, we can get
$$
\sigma_{DA}^2=\frac{\mu^2\delta}{\alpha^2}=20,\ \tilde{\Delta_{DA}}=\sqrt{\frac{\sigma_{DA}^2}{2\delta}}=5\sqrt{2}=7.0711
$$

```r
#The answer to question(e) is as follows.
Rou_DA <- function(Delta)
{
    DeltaWave_DA <- sqrt(mu^2/(2*alpha^2))
    rou_DA_value <- alpha/delta*(1-exp(-Delta/DeltaWave_DA))
    return(rou_DA_value)
}
```

```r
#The answer to question(f) is as follows.
logging <- F  #To speed up the simulation

alpha <- 1
mu <- 10
delta <- 1/5
numEvents <- 100000  # Average over 100,000 events
initializeBook5()

# Burn in for 100 events
for(count in 1:100){
    generateEvent()
}

avgBookShape <- bookShape(29)/numEvents
for(count in 2:numEvents){
    generateEvent()
    avgBookShape <- avgBookShape+bookShape(29)/numEvents
}

plot(-29:29,avgBookShape,main=NA,xlab="Relative price",ylab="# Shares", col="red", type="b")
lines(x <- c(0:29),y <- Rou_DA(x),type = "l",col = "blue")  #The theoretical asymptotic curve
lines(x <- c(10:29),y <- rep(mean(avgBookShape[x+30]),20),type = "l",col = "orange")  #The practical asymptotic line
```
